```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
import os
from bs4 import BeautifulSoup as bs
import shutil
```

```python
def initialiseChromeDriver():
    driver = webdriver.Chrome(executable_path= r'C:\Program Files\chromedriver\chromedriver')
    driver.get("https://www.amazon.in/ref=nav_logo")
    return driver

def fill_form(driver, search_argument):
    '''Finds all the input tags in form and makes a POST requests.'''
    search_field = driver.find_element_by_id('twotabsearchtextbox')
    search_field.send_keys(search_argument)
    # We look for the search button and click it
    driver.find_element_by_class_name('nav-input')\
    .click()

#creating directory
def make_directory(dirname):
    current_path= os.getcwd()
    path= os.path.join(current_path, dirname)
    if not os.path.exists(path):
        os.makedirs(path)

def startScrapingContent(driver):
    ##beautiful soup##
    html = driver.page_source #html page source
    soup = BeautifulSoup(html)
    
    urls = []
    
    for tag in soup.find_all("a", {"class":"a-link-normal"}):
        img = tag.find_all('img')
        #print (str(len(img)) + "images found.")
        for i in img:
            link = i['src']
            urls.append(link)
        
        
    #navigating to next page
    driver.find_element_by_class_name('a-normal')\
    .click()
   
    #Adding next page image urls to URLs
    for tag in soup.find_all("a", {"class":"a-link-normal"}):
        img = tag.find_all('img')
        #print (str(len(img)) + "images found.")
        for i in img:
            link = i['src']
            urls.append(link)
    
    #saving images
    for index,link in enumerate(urls):
        print("downloading {0} of {1} images".format(index+1, len(urls)))
        response= requests.get(link)
        with open('img_{0}.jpeg'.format(index),"wb") as file:
            file.write(response.content)

driver = initialiseChromeDriver()
fill_form(driver, 'women sarees')
make_directory('sarees')
os.chdir("C:\\Users\\preetham\\Data Science\\Webscrapper Amazon\\sarees")
startScrapingContent(driver)
driver.close()
```
